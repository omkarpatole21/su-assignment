Q5. Plot two line graphs on the same figure:

y1 = [2, 4, 6, 8, 10]

y2 = [1, 3, 5, 7, 9]

Use different colors and add a legend.

```python
x = [1,2,3,4,5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
```

```python
plt.plot(x,y1, color='blue', marker='o')
plt.plot(x,y2, color='red', marker='o')
plt.title('Two Line Graphs on the Same Figure')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

# Seaborn-Line and Bar Plots

Q13. Using Seaborn, plot a line chart for the dataset:

import seaborn as sns

```python
data = {'Day': [1,2,3,4,5], 'Temperature': [30,32,34,33,35]}
df=pd.DataFrame(data)
```
