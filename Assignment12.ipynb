## 3)DataFrame Merging

Q17. Create two DataFrames and merge them using pd.concat.

```python
df1=pd.DataFrame({'ID':[1,2,3],'Name':['Sahil','Om','Ashish']})
```

```python
df2=pd.DataFrame({'ID':[1,2,4],'Salary':[20000,45000,34000]})
```

```python
pd.concat([df1,df2])
```

Q18. Merge two DataFrames using a common column with pd. merge.

```python
pd.merge(df1,df2, on='ID')
```

Q19. Perform an inner join and outer join between two DataFrames.

```python
pd.merge(df1,df2, on='ID', how='inner')
```

```python
pd.merge(df1,df2, on='ID', how='outer')
```
